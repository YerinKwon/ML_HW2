# Data Analysis & Preprocessing
## Basic Reshaping
원본 파일의 오타,주석 등을 처리하여 올바른 csv 파일로 변경   

```python
data_path = 'Chronic_Kidney_Disease/data.csv'
data = pd.read_csv(data_path)
```

텍스트로 된 범주형 데이터 숫자로 인코딩   

```python
categorized = ['sg','al','su','rbc','pc','pcc','ba','htn','dm','cad','appet','pe','ane','class']
continuous = ['age','bp','bgr','bu','sc','sod','pot','hemo','pcv','wbcc','rbcc']
```

```python
for column in data.columns:
    if column in categorized:
        encoded, category = data[column].factorize()
        data[column] = encoded
        data[column].replace(-1,np.NaN, inplace=True)
```

## Missing Data

```python
import missingno as msno
```

```python
msno.matrix(data)
```

```python
data.dropna(axis=1, thresh=300,inplace=True)
```

## Preprocessing
1. 범주형 데이터의 결측값 : 최빈값으로 대체   
2. 연속형 데이터의 결측값 : 평균값으로 대체   

```python
for column in data.columns:
    if column in categorized:
        data[column].replace(np.NaN,data[column].mode()[0], inplace=True)
    else:
        data[column].replace(np.NaN, data[column].mean(),inplace=True)
```

```python
data
```

# Classification

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
ckd_data = data.loc[data['class'] == 0]
notckd_data = data.loc[data['class'] == 1]
```

```python
model = RandomForestClassifier()
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.iloc[:,:-1], data.iloc[:,-1], test_size=0.5, random_state=2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
y_predicted = model.predict_proba(x_test)[:,1]
```

```python
FP, TP, thresholds= roc_curve(y_test,y_predicted)
auc = roc_auc_score(y_test, y_predicted)
```

```python
plt.figure()
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(FP, TP, color='darkorange', label='Random Forest')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve (area = %0.2f)' % auc)
plt.legend(loc='best')
plt.show()
```
